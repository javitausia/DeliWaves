```python
import glob
import pandas as pd
```

```python
for i, file in enumerate(sorted(glob.glob('*.pkl'))):
    data = pd.read_pickle(file)
    data = data[['Hs_Agg','Tp_Agg','Dir_Agg']].rename(columns={'Hs_Agg': 'Hs_{}'.format(file[21:23]),
                                                               'Tp_Agg': 'Tp_{}'.format(file[21:23]),
                                                               'Dir_Agg': 'Dir_{}'.format(file[21:23])})
    if i==0:
        laredo = data
    else:
        laredo = laredo.join(data)
```
